```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.arange(-4, 4, 0.02)
# a = [-3, -2 , -1 ,0, 1, 2, 3]
# b = [1, 2, 3, 4]
b = a

a, b = np.meshgrid(a, b)

print(a)
print(b)
```

```python
m = a.shape[0]
n = a.shape[1]
print(m, n)

for i in range(m):
    for j in range(n):
#         print("(%d, %d)"%(b[i, j], a[i, j]), end="")
    print()
```

```python
fig = plt.figure()
```

```python
axes = fig.gca(projection='3d')
axes.plot_surface(a, b, a**2+b**2, cmap='rainbow')
plt.show()
```

```python
fig = plt.figure()
axes = fig.gca(projection='3d')
axes.contour(a, b, a**2+b**2, cmap='rainbow')
plt.show()
```
