# Generating Regression Data

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
u = 0
std = 1

x = (np.random.randn(500)*std  + u)
noise = np.random.randn(500)
# print(x)
```

```python
plt.hist(x)
plt.show()
```

```python
y = 3*x + 2 + noise
print(x.shape, y.shape)
```

```python
plt.scatter(x, y, color='red')
plt.show()
```

# Data for Classification

Let us generate some examples for Binary Classification

```python
u1 = np.array([2, 4])
cov_1 = np.array([[2, -.8], 
                  [-.8, 1]])

u2 = np.array([-1, -3])
cov_2 = np.array([[1, 0], 
                  [0, 2]])
```

```python
X1 = np.random.multivariate_normal(u1, cov_1, 300)
print(X1.shape)

X2 = np.random.multivariate_normal(u2, cov_2, 400)
print(X2.shape)
```

```python
plt.scatter(X1[:, 0], X1[:, 1], label="Apple")
plt.scatter(X2[:, 0], X2[:, 1], label="Orange")
plt.title("Apples and Oranges")
plt.xlabel("X1 (size)")
plt.ylabel("X2 (sweetness)")

plt.show()
```

```python
y = np.zeros(700)
y[:300] = 1
print(y.shape)
print(y)
```

```python
# Vertical Stacking the Matrices [[300x2][400x2]] -> [700x2]

X = np.vstack((X1, X2))
print(X.shape)
```

```python
# plt.scatter(X[:, 0], X[:, 1])

for i in range(700):
    if y[i] == 1:
        plt.scatter(X[i, 0], X[i, 1], color='green')
    else:
        plt.scatter(X[i, 0], X[i, 1], color='red')
    
plt.show()
```

# Shortcut

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
# c=y assigns new color for each class
plt.show()
```
