# Generating Datasets

Some useful functions to generate datasets available in `sklearn.datasets`
- make_regression
- make_classification
- make_blobs
- make_moons

```python
from matplotlib import pyplot as plt

from sklearn.datasets import make_regression, make_classification, make_blobs, make_moons
```

```python
X, Y = make_regression?
```

```python
X, Y = make_regression(n_samples=400, n_features=1, n_informative=1, noise=10.8)
```

```python
print(X.shape, Y.shape)
```

```python
plt.scatter(X, Y)
plt.show()
```

```python
make_classification?
```

```python

XC, YC = make_classification(n_samples=400, n_features=2, n_clusters_per_class=1, n_classes=3 ,n_informative=2, n_redundant=0, random_state=5)
```

```python
print(XC.shape, YC.shape)
```

```python
plt.scatter(XC[:, 0], XC[:, 1], c=YC)
plt.show()
```

# Blobs Dataset

```python
XB, YB = make_blobs?

# For CLUSTERING
```

```python
XB, YB = make_blobs(n_samples=1000, n_features=2, centers=5, random_state=10)
```

```python
plt.scatter(XB[:, 0], XB[:, 1], c=YB)
plt.show()
```

# Make Moons Dataset

```python
make_moons?
```

```python
Xm, Ym = make_moons(n_samples=100, noise=0.2, random_state=5)
```

```python
plt.scatter(Xm[:, 0], Xm[:, 1], c=Ym)
plt.show()

# USEFUL FOR Non Linear DATASETS
```

# Subplots

```python
plt.figure(figsize=(8,8))

plt.subplot(221)
plt.title("Regression")
plt.scatter(X, Y)

plt.subplot(222)
plt.title("Classification")
plt.scatter(XC[:, 0], XC[:, 1], c=YC)

plt.subplot(223)
plt.title("Blobs")
plt.scatter(XB[:, 0], XB[:, 1], c=YB)

plt.subplot(224)
plt.title("Moons")
plt.scatter(Xm[:, 0], Xm[:, 1], c=Ym)

plt.show()
```
