# Data Visualisations

### Install matplotlib

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(10)
x = x**2
x2 = x**3
print(x)
```

```python
plt.plot(x,color='red')
plt.show()
plt.plot(x2)
plt.show()
```

```python
plt.scatter(x, x2, marker="^", color="orange", label="stock price")
plt.legend()
plt.xlabel("days")
plt.ylabel("price")
plt.title("Stock Price Prediction")
plt.show()
```

```python
plt.plot(x,x**2, label='2018')
plt.plot(x,x**3, label='2017')
plt.xlabel("days")
plt.ylabel("price")
plt.title("Stock Price Prediction")
plt.legend()
plt.show()
```

# Bar Charts

```python
# 0, 10 are the range and 5 is the size
numbers = np.random.randint(1, 10, 5)
indices = np.arange(5)

numbers2 = np.random.randint(1, 10, 5)
indices2 = np.arange(5)

print(numbers)
print(indices)
```

```python
plt.bar(indices, numbers, 0.25, color='red', label = 'Gold')
plt.bar(indices2+0.25, numbers2, 0.25, color='green', label = 'silver')
plt.legend()
plt.show()
```

# Pie Chart

```python
labels = ['Algebra', 'Calculus', 'Coding', 'Algorithms']
values = [90, 80, 54, 40]

plt.pie(values, labels = labels, radius = 2, explode=(0, 0, 1, 0), shadow = True)
plt.show()
```

# Histogram

```python
# Generate a Normal Distribution (1-D)
```

```python
u = 5
sigma = 1
# Such is also called as Standard Normal Distribution

vals = u + sigma*np.random.randn(200)

print(vals.shape)

plt.hist(vals, 10)
plt.show()
```

```python
vals = np.round(vals)
z = np.unique(vals, return_counts=True)
print(z)

# Numbers & Frequency (How many numbers are closer to the number 5)

## Rounding off values -> Ignore
```

```python
x = vals
y = np.zeros(x.shape)

plt.scatter(x, y)
plt.show()
```

# Multivariate Normal Distribution

```python
# BiVariate
mean = np.array([0.0, 0.0])
cov = np.array([[1,0.8], 
                [0.8,1]])

mean2 = np.array([5.0, 6.0])
cov2 = np.array([[1.3,0.2], 
                [0.2,1.1]])


dist = np.random.multivariate_normal(mean, cov, 500)
print(dist.shape)

dist2 = np.random.multivariate_normal(mean2, cov2, 500)
print(dist2.shape)

plt.scatter(dist[:, 0], dist[:, 1])
plt.scatter(dist2[:, 0], dist2[:, 1])
plt.show()
```
